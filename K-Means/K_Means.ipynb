```python
features, true_labels = make_blobs(n_samples=200, centers=3, cluster_std=2.75, random_state=42)
```

```python
features
```

```python
true_labels
```

```python
plt.scatter(features[:,0], features[:,1], c=true_labels, cmap='rainbow')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
```

```python
scaled_features = scaler.fit_transform(features)
```

```python
print('Features: \n', scaled_features[:5])
```

```python
kmeans = KMeans(init='random', n_clusters=3, n_init=10, max_iter=300, random_state=42)
```

```python
kmeans.fit(scaled_features)
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
plt.scatter(scaled_features[:,0], scaled_features[:,1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red', marker='*', label='centroids')
plt.legend()
plt.show()
```
